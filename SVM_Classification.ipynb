```python
from sklearn.datasets import fetch_20newsgroups

twentytrain=fetch_20newsgroups(subset='train',shuffle=True)
```

```python
twentytrain.keys()
```

```python
print(twentytrain.data[0])
```

```python
twentytrain.target_names
```

we need to support our numeric data into a vector name

using count vectorizer to count the words frequency along with the word id and document id 


```python
from sklearn.feature_extraction.text import CountVectorizer 

count_vec=CountVectorizer()


X_train_counts=count_vec.fit_transform(twentytrain.data)

X_train_counts.shape
```

```python
print(X_train_counts[0])
```

tfidf vectorizer

```python
from sklearn.feature_extraction.text import TfidfTransformer


Tfidf_Transformer=TfidfTransformer()


X_train_tfidf=Tfidf_Transformer.fit_transform(X_train_counts)

X_train_tfidf.shape
```

penalty is being given as l2 norm and tolenrence has been provided to stop training sets on our model just to see if we our objective function below this tolerence level we will stop training our model to avoid overfitting of our model



```python
from sklearn.svm import LinearSVC

clf_svm=LinearSVC(penalty="l1", dual=False,tol=1e-3)
clf_svm.fit(X_train_tfidf,twentytrain.target)
```

```python
from sklearn.pipeline import Pipeline

clf_svc_pipeline=Pipeline([
    ('vect',CountVectorizer()),
    ('tfidf',TfidfTransformer()),
    ('clf',LinearSVC(penalty='l1',dual=False,tol=0.001))
])
```

```python
clf_svc_pipeline.fit(twentytrain.data,twentytrain.target)
```

importing test data now 

```python
twentytest=fetch_20newsgroups(subset='test',shuffle=True)
```

```python
predicted=clf_svc_pipeline.predict(twentytest.data)
```

```python
from sklearn.metrics import accuracy_score


acc_svm=accuracy_score(twentytest.target,predicted)

acc_svm
```
