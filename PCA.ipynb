```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes=load_diabetes()
```

```python
X=diabetes['data']
```

```python
X.shapeP
```

mean centering

```python
# taking mean of all column values
mu=X.mean(axis=0)

# mean centering

X_=(X-mu)
```

covariance matrix

```python
S=np.dot(X_.T,X_)
S.shape
```

getting eigan value decomposition(getting eigan value and eigan vector)

```python
lambdas,vectors=np.linalg.eig(S)
```

```python
lambdas.shape
```

```python
vectors.shape
```

to get top k eigan vectors we have extracted the values

```python
v=vectors[:,:2]
```

```python
v.shape
```

Projection of the Data

```python
X_.shape
```

```python
v.shape
```

we have reduced the dimnesion from 10 to just 2 dimensions

```python
X_new=np.dot(X_,v)
```

```python
X_new.shape
```

```python
X_new
```

```python
total=sum(lambdas)
```

```python
explanined_variance=np.round(lambdas/total,3)
explanined_variance
```

```python
cummvaltive_variance=np.cumsum(explanined_variance)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(cummvaltive_variance,marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Variance Explained")
plt.grid("on")
```
