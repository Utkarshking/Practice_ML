```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

data normalisation

```python
def normalise(X):
    u=X.mean(axis=0)
    std=X.std(axis=0)
    return X-u/std
X=normalise(X)
```

```python
# intialising k centres for k clusters 
k=5
colors=["green","red","blue","orange","yellow"]
n_features=2
centroids={}
for i in range(k):
    # we are multiplying the range number by 2 and subtracting it by 1 to give the range from -1,1
    # as np.random.random function will always provide us the range from (0,1)
    center=2*(2*np.random.random((n_features,))-1)
    centroids[i]={
        'center':center,
        'color':colors[i],
        'points':[]
    }
```
