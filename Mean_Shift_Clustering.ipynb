K Means Clustering :-


Need to specify the number of clusters as hyperparameter
Cant handle some non linear complex data 
Less Hyperparameter tuning needed
Computationally very intesive
Performs at almost O(N) data points where n is the number of data points

struggles with outliers

Mean Shift Clustering :-


No Need to specify number of clusters upfront as hyperparameters 

Uses Density function to handle the non linear complex data (eg pixels)


hyperparamter tuning is required


Computationally very extensive

performs at almost O(N^2) data points where n is the number of data points


Kernels:-


Kernels are a powerful tool in machine learning, especially in clustering algorithms, that can help to capture complex relationships between data points

They provide a way to map data into a higher-dimensional space where it might be easier to separate clusters


How Kernels Work:

Feature Space: Kernels implicitly map the original data points into a higher-dimensional feature space.
Similarity Measure: In this higher-dimensional space, the similarity between data points is computed using a kernel function.
Clustering: The clustering algorithm then operates on these similarity measures to group data points




```python
import pandas as pd

wine_clustering=pd.read_csv('wine-clustering.csv')

wine_clustering.head()
```

```python
wine_clustering.info()
```

Instanititating mean shift clustering for the model

bandwidth is being passed as a hyperparamter to the meanshift initialisation of the model

```python
from sklearn.cluster import MeanShift

analyser=MeanShift(bandwidth=150)


analyser.fit(wine_clustering)

```

you don't need to guess the bandwidth as a hyperparameter which can also be done sklearn algorithm

```python
from sklearn.cluster import estimate_bandwidth


estimate_bandwidth(wine_clustering)
```

getting the labels from the meanshift clustering model

```python
labels=analyser.labels_
```

```python
import numpy as np


np.unique(labels)
```

```python
wine_clustering['ClusterGroup']=np.nan

for i in range(len(wine_clustering)):
    wine_clustering.iloc[i,wine_clustering.columns.get_loc('ClusterGroup')]=labels[i]
```

```python
wine_clustering.head()
```

```python
wine_clustering_data=wine_clustering.groupby(['ClusterGroup']).mean()
wine_clustering_data
```

```python
wine_clustering_data['Counts']=pd.Series(wine_clustering.groupby(['ClusterGroup']).size())

wine_clustering_data
```
