```python
import cv2
import matplotlib.pyplot as plt
```

reading images

```python
img=cv2.imread("kids.jpg")
print(img.shape)

devimg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
plt.imshow(devimg)
plt.axis("off")
plt.show()
```

initalising the model for the frontalface convolutional detection

```python
model=cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

detect multiscale function is used to detect the objects through a particular template in an image

```python
faces=model.detectMultiScale(devimg,1.7,5)
```

the faces have coordinates which are [x,y,height,width]

```python
faces
```

```python
for face in faces:
    x,y,w,h=face
    cv2.rectangle(devimg,(x,y),(x+w,y+h),(0,255,0),2)


plt.imshow(devimg)
```
