```python
import pandas as pd
pd.__version__
```

```python
laptop_data=pd.read_csv("dataset.csv")
laptop_data
```

```python
def data_preprocess(data , x):
    temp = pd.get_dummies(data[x],drop_first = True).astype(int)
    data = pd.concat([data,temp],axis =1)
    data = data.drop(columns =x)
    return data
    
df = data_preprocess(laptop_data,'Company')
df =data_preprocess(laptop_data,'TypeName')
df = data_preprocess(laptop_data,'CPU_Company')
df = data_preprocess(laptop_data,'GPU_Company')
df = data_preprocess(laptop_data,'OpSys')
df
```

```python
X = df.drop(columns= 'Price (Euro)') # Features
Y = df['Price (Euro)'] # Target column
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder


label=LabelEncoder()

for columns in X.select_dtypes('object').columns:
    X[columns]=label.fit_transform(X[columns])
```

Numerical Normalisation

```python
Numerical_columns = df.select_dtypes(exclude = 'object').columns



from sklearn.preprocessing import MinMaxScaler


scaler=MinMaxScaler()


df[Numerical_columns]= scaler.fit_transform(df[Numerical_columns])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split


X_train,X_Test,Y_Train,Y_Test=train_test_split(X,Y,test_size=0.2)

linear_model=LinearRegression()

linear_model.fit(X_train,Y_Train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
Y_Pred = linear_model.predict(X_Test)

MSE = mean_squared_error(Y_Test,Y_Pred)
r2 = r2_score(Y_Test, Y_Pred)

MSE,r2
```

```python
linear_model.score(X_train,Y_Train)
```

```python
predictors=X_train.columns

coef=pd.Series(linear_model.coef_,predictors).sort_values()


coef
```

```python
import matplotlib.pyplot as plt

plt.plot(Y_Pred,label='Predicted')
plt.plot(Y_Test,label='Actual')


plt.ylabel('Price')


plt.legend()


plt.show()
```

Gradient Boosting Regressor

```python
from sklearn.ensemble import GradientBoostingRegressor


params={'n_estimators':500,'max_depth':6,'min_samples_split':2,'learning_rate':0.01,'loss':'squared_error'}

gbr_model=GradientBoostingRegressor(**params)


gbr_model.fit(X_train,Y_Train)
```

```python
gbr_model.score(X_train,Y_Train)
```

```python
Y_Pred=gbr_model.predict(X_Test)
```

```python
plt.rcParams['figure.figsize']=(15,6)

plt.plot(Y_Pred,label='Predicted')

plt.plot(Y_Test.values,label='Actual')

plt.legend()

plt.show()
```

Using grid search to find the best possible model

```python
from sklearn.model_selection import GridSearchCV

num_estimators=[100,200,500]

learn_rates=[0.01,0.02,0.05,0.1]
max_depths=[4,6,8]

param_grid={
    'n_estimators':num_estimators,
    'learning_rate':learn_rates,
    'max_depth':max_depths
}


grid_search=GridSearchCV(GradientBoostingRegressor(min_samples_split=2,loss='squared_error'),param_grid,cv=3,return_train_score=True)


grid_search.fit(X_train,Y_Train)


grid_search.best_params_
```

```python
len(grid_search.cv_results_)
```

```python
for i in range(36):
    print('paramater',grid_search.cv_results_['params'][i])
    print('Mean Test Score', grid_search.cv_results_['mean_test_score'][i])
    print('Rank : ',grid_search.cv_results_['rank_test_score'][i])
    print()
```

```python
params={
   'learning_rate': 0.1, 'max_depth': 4, 'n_estimators': 500,'loss':'squared_error','min_samples_split':2
}

gbr_model=GradientBoostingRegressor(**params)

gbr_model.fit(X_train,Y_Train)
```
